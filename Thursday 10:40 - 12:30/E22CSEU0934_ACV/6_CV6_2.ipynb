```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('carsroad.jpg', cv2.IMREAD_GRAYSCALE)

edges = cv2.Canny(image, 100, 200)

plt.figure(figsize=(10,5))
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(edges, cmap='gray'), plt.title('Canny Edge Detection')
plt.show()
```

```python

lines = cv2.HoughLinesP(edges, 1, np.pi/180, 68, minLineLength=15, maxLineGap=250)

# Draw detected lines
line_image = np.copy(image)
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(line_image, (x1, y1), (x2, y2), (0, 255, 0), 2)

plt.imshow(line_image)
plt.title('Detected Lines using Hough Transform')
plt.show()
```

```python
import torch
import torchvision
import torch.nn as nn
import torchvision.transforms as transforms
import torchvision.models as models
import matplotlib.pyplot as plt
import numpy as np
from skimage.segmentation import felzenszwalb
from torch.utils.data import DataLoader
from torchvision.datasets import FashionMNIST


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((224, 224))  # Resize for ResNet input
])

train_dataset = FashionMNIST(root="./data", train=True, transform=transform, download=True)
train_loader = DataLoader(train_dataset, batch_size=1, shuffle=True)

cnn_model = models.resnet18(pretrained=True)
cnn_model.fc = nn.Identity()  # Remove classification layer, we only need features
cnn_model = cnn_model.to(device)
cnn_model.eval()


image, label = next(iter(train_loader))
image = image.to(device)

# Convert tensor to numpy image for region segmentation
np_image = image.squeeze(0).permute(1, 2, 0).cpu().numpy()

# Apply Felzenszwalb segmentation (Selective Search alternative)
segments = felzenszwalb(np_image, scale=100, sigma=0.5, min_size=500)
regions = []
for i in np.unique(segments):
    y, x = np.where(segments == i)
    x1, y1, x2, y2 = np.min(x), np.min(y), np.max(x), np.max(y)
    regions.append((x1, y1, x2, y2))

# Extract features for top 10 regions
features = []
for x1, y1, x2, y2 in regions[:10]:  # Limit to 10 regions for speed
    roi = image[:, :, y1:y2, x1:x2]  # Crop ROI
    roi = transforms.functional.resize(roi, (224, 224))  # Resize for CNN

    # Ensure correct shape: [Batch, Channels, Height, Width]
    if roi.ndim == 3:
        roi = roi.unsqueeze(0)  # Add batch dimension

    # Convert grayscale (1 channel) to 3-channel RGB
    roi = roi.expand(-1, 3, -1, -1)  # Efficient way to repeat across channels

    with torch.no_grad():
        feature = cnn_model(roi.to(device))
    features.append(feature.squeeze().cpu().numpy())

# Display the segmented regions
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(np_image.squeeze(), cmap="gray")
ax[0].set_title("Original Image")
ax[1].imshow(segments, cmap="nipy_spectral")
ax[1].set_title("Segmented Regions")
plt.show()

print("Extracted features shape for each region:", features[0].shape)
```
