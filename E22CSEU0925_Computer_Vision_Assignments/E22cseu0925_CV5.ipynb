```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_score, recall_score, f1_score
import seaborn as sns
import struct

# Task 1: Image Compression
def compress_image(image_path, output_path, compression_type="jpeg", quality=50):
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Unable to load image. Check the path.")
        return
    if compression_type.lower() == "jpeg":
        cv2.imwrite(output_path, image, [cv2.IMWRITE_JPEG_QUALITY, quality])
    elif compression_type.lower() == "png":
        cv2.imwrite(output_path, image, [cv2.IMWRITE_PNG_COMPRESSION, 9])
    else:
        print("Unsupported compression type. Use 'jpeg' or 'png'.")

# Example usage
compress_image("/content/I1.jpg", "/content/I1_compressed.jpg", "jpeg", 50)

# Task 2: CNN on MNIST Dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
split_idx = int(0.8 * len(x_train))
x_train, x_val = x_train[:split_idx], x_train[split_idx:]
y_train, y_val = y_train[:split_idx], y_train[split_idx:]

# Define CNN model
model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation="relu", input_shape=(28, 28, 1)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation="relu"),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dense(10, activation="softmax")
])

# Compile model
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=50, batch_size=64)

test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy (MNIST): {test_acc:.4f}")
```
